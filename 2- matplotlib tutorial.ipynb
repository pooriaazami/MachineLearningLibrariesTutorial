# Basic Plotting

## plot

```python
x = np.linspace(-10, 10, 1000)
y = x ** 2

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-10, 10, 1000)

plt.plot(x, x ** 2)
plt.plot(x, x ** 3)
plt.show()
```

## scatter

```python
x = np.array([1, 2, 3])
y = np.array([1.5, 2.5, 3.5])

plt.scatter(x, y)
plt.show()
```

## color, marker

```python
x = np.linspace(-10, 10, 1000)
y = x ** 2

plt.plot(x, y, color='black')
plt.show()
```

```python
x = np.linspace(-10, 10, 1000)

plt.plot(x, x ** 2, color='black')
plt.plot(x, x ** 3, color='green')

plt.show()
```

```python
x = np.arange(10)
y = x ** 2

plt.plot(x, y, marker='+')

plt.show()
```

## legend

```python
x = np.linspace(-10, 10, 1000)

plt.plot(x, x ** 2, label='x^2')
plt.plot(x, x ** 3, label='x^3')

plt.legend()
plt.show()
```

## title

```python
x = np.arange(10)
y = x ** 2

plt.plot(x, y, marker='+')
plt.title('x ^ 2')

plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Subplots

```python
fig, axs = plt.subplots(4, 4, figsize=(15, 15))

x = np.linspace(-10, 10, 1000)

for i in range(4):
    for j in range(4):
        axs[i, j].plot(x, x ** (i * 2 + j))

plt.show()
```
