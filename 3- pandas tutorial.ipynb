# Pandas Introduction

pandas is a widely used library for manipulating semi-structured and structured files. These files can have many different formats ranging from SQL databases to CSV files.

The most commonly used format for saving semi-structured data is `.csv` which is a short form of comma-separated values, which means that every row of the file contains a tuple of data, and each value in that tuple is separated with a comma.

As an example consider the file below:

```
name,family,age
Pooria,Azami,22
```

This file represents the below table:

|name|family|age|
|:-------------:|:------------:|:-------------:|
|Pooria|Azami|22|

The very first line of the files is called the `header`. 
Note that it is possible to separate data within the tuples with something other than a comma, like a tab, a semicolon, and many others.

Visit [this link](https://pandas.pydata.org/) to explore tge library.

# Loading A File

````python
DATASET_PATH = 'data.csv'
````

To load a CSV file with pandas you can use the `read_csv` method. The first argument of the method is the dataset path. You can also pass the `sep` parameter to tell pandas that it should separate tuples with something other than commas. It is also possible to tell pandas that the file does not have a header by setting the `header` parameter to `None`.

````python
df = pd.read_csv(DATASET_PATH) # df stands for DataFrame
````

To see top 5 tuples of the dataframe we can use `.head()` method.

````python
df.head()
````

We can also pass the count of tuples that we want it to show.

````python
df.head(10)
````

There is also a `.tail()` method that shows last tuples of the dataframe.

````python
df.tail()
````

# Basic File Exploring

`.describe()` is a useful method that shows the basic statistics of the file.

````python
df.describe()
````

You can list the name of the dataframe columns with the code below:

````python
for name in df.columns:
    print(name, end='\t') # Note that I have used the `end` parameter to have a cleaner output
````

Tp get the shape of the dataframe you can use the `.shape` attribute.

````python
print(f'df.shape: {df.shape}\n')

count_data, count_columns = df.shape
print(f'There are {count_data} tuples in this datafram.')
print(f'Every tuple has {count_columns} items.')
````

# Accessing Tha Data

You can use `.iloc` to select specific rows of the dataframe.

````python
df.iloc[0]
````

````python
df.iloc[[0, 10, 20]]
````

To select columns of the data you can use a dictionary like indexing.

````python
df['diagnosis']
````

````python
df[['diagnosis', 'radius_mean']]
````

To remove some columns you can use the `.drop` method.

````python
new_df = df.drop(columns=['id', 'Unnamed: 32'])
new_df.head()
````

To remove columns from the actual dataframe, set the `inplace` parameter to `True`.

````python
df.drop(columns=['id', 'Unnamed: 32'], inplace=True)
````
